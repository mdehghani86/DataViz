### Understanding the Anatomy of a figure in python

Matplotlib library is built upon Numpy. It uses a tree−like hierarchical structure which consists of objects that makeup each of these plots.

A ’Figure’ in Matplotlib can be understood as the outermost storage for a graph. This ‘Figure’ can contains multiple ‘Axes’ objects. ‘Axes’ object is NOT the plural form of ‘Axis’ in this case.

‘Axes’ can be understood as a part of ‘Figure’, a subplot. It can be used to manipulate every part of the graph inside it. A ‘Figure’ object in an Matplotlib is a box that stores one or more ‘Axes’ objects.

Under ‘Axes’ comes the tick marks, lines, legends, and text boxes in the hierarchy. Every object in the Matplotlib can be manipulated.

# 1. Line Chart 



- Bivariate Analysis
- categorical -> numerical and numerical -> numerical
- Use case - Time series data

```python
# Creating a Sample Dataframe using Dict and Pandas DataFrame function

data = { 'price':[48000,54000,57000,49000,47000,45000],
        'year' : [2015,2016,2017,2018,2019,2020] }

df = pd.DataFrame(data)
```

```python
# Using the Plot function to plot the line chart

plt.plot(df['year'],df['price'])
```

```python
# Creating another sample dataframe 

data = { 'product' : ['Bicycle','Bicycle','Bicycle','Bicycle','Bicycle','Bicycle','Mobile','Mobile','Mobile','Mobile','Mobile','Mobile'],
    'price':[4800,5400,5700,4900,4700,4500,8000,9000,9500,11000,15000,16000],
        'year' : [2015,2016,2017,2018,2019,2020,2015,2016,2017,2018,2019,2020] }

df = pd.DataFrame(data)
df
```

```python
# plotting multiple plots
# reshape the data using pivot_table
# Pandas Pivot_table -- https://pandas.pydata.org/docs/reference/api/pandas.pivot_table.html

df_pivot = df.pivot_table(values='price', index='year', columns='product')

df_pivot
```

```python
# Using implicit method 
# Implicit coding refers to using the default behavior of Matplotlib to create plots. 
# For example, if you call the plot() function with a list of values, Matplotlib will create a line plot using those values

ax = df_pivot.plot()
```

```python
# Using Explict Method
# Refers to specifying all the parameters of the plot explicitly. 
# This can be useful for customizing the plot or creating more complex visualizations.

fig, ax = plt.subplots()
ax.plot(df_pivot.index, df_pivot['Bicycle'])
ax.plot(df_pivot.index, df_pivot['Mobile'])
```

```python
# Adding Title and Axes Labels

ax = df_pivot.plot()
ax.set(title='Sales by Year', xlabel='Year', ylabel='Sales')
```

```python
# colors(hex) and line(width and style) and marker(size)
# Which Method of calling the chart is used here ?

plt.plot(df_pivot.index,df_pivot['Bicycle'],color='#D9F10F')
plt.plot(df_pivot.index,df_pivot['Mobile'],color='#FC00D6')

plt.title('Sales by Year')
plt.xlabel('Year')
plt.ylabel('Sales')
```

```python
# Line Style

plt.plot(df_pivot.index,df_pivot['Bicycle'],color='#D9F10F',linestyle='solid',linewidth=3)
plt.plot(df_pivot.index,df_pivot['Mobile'],color='#FC00D6',linestyle='dashdot',linewidth=2)

plt.title('Sales by Year')
plt.xlabel('Year')
plt.ylabel('Sales')
```

```python
# Marker Style

plt.plot(df_pivot.index,df_pivot['Bicycle'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(df_pivot.index,df_pivot['Mobile'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Sales by Year')
plt.xlabel('Year')
plt.ylabel('Sales')
```

```python
# legend -> location
plt.plot(df_pivot.index,df_pivot['Bicycle'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10,label='Bicycle')
plt.plot(df_pivot.index,df_pivot['Mobile'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o',label='Mobile')

plt.title('Sales by Year')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.legend(loc='upper right')
```

```python
# limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.ylim(0,75000)
plt.xlim(2017,2019)
```

```python
# Plotting Grid

plt.plot(df_pivot.index,df_pivot['Bicycle'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(df_pivot.index,df_pivot['Mobile'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Sales by Year')
plt.xlabel('Year')
plt.ylabel('Sales')

plt.grid()
```

### Bar chart

image.png

- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
# Creating a Sample Dataframe

time_data = {'facebook':120, 'youtube':90, 'instagram':40, 'snapchat':28}
social_media = list(time_data.keys())
time_spend = list(time_data.values())
```

```python
# Using Style and Color Palette

style.use('ggplot')
plt.figure(figsize=(5,5))

color = ['royalblue', 'red', 'purple', 'yellow']
plt.bar(social_media, time_spend, color=color, width=0.5)

plt.xlabel('Social Media')
plt.ylabel('Time in Minutes')
plt.title('Average time spend on social media')
```

```python
# Flipping the Axis

plt.barh(social_media, time_spend, color=color)
plt.xlabel('Social Media')
plt.ylabel('Time in Minutes')
plt.title('Average time spend on social media')
```

```python
#Creating a Sample Dataframe

fb = (120, 90, 110, 100, 95)
yt = (80, 90, 75, 85, 95)
ig = (50, 40, 55, 35, 45)
sc = (20, 30, 35, 25, 28)
xpos = np.arange(len(fb))
xpos
```

```python


style.use('ggplot')
plt.figure(figsize=(10,7))
barWidth = 0.2
plt.bar(xpos, fb, color='royalblue', width = barWidth, label='FB')
plt.bar(xpos+0.2, yt, color='red', width = barWidth, label='YT')
plt.bar(xpos+0.4, ig, color='purple', width = barWidth, label='IG')
plt.bar(xpos+0.6, sc, color='yellow', width = barWidth, label='SC')
plt.xticks(xpos+0.3, ('Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'))
plt.xlabel('Day')
plt.ylabel('Time spend in minutes')
plt.title('Time spend on social media')
plt.legend()
```

```python
fb = (120, 90, 110, 100, 95)
yt = (80, 90, 75, 85, 95)
ig = (50, 40, 55, 35, 45)
x = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
xpos = np.arange(len(x))
barWidth = 0.4
plt.figure(figsize=(10,7))
plt.bar(xpos, fb, color='royalblue', width= barWidth, label='FB')
plt.bar(xpos, yt, bottom=fb, color='red', width= barWidth, label='YT')
plt.bar(xpos, ig, bottom=yt, color='purple', width= barWidth, label='IG')
plt.xlabel('Day')
plt.ylabel('Time spend in minutes')
plt.title('Time spend on socia media')
plt.xticks(xpos, x)
plt.legend()
```

### Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# Creating sample Dataframe and Ploting using Pyplot.Pie()

exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
```

```python
# Pie chart as perfect circle
# Plt.axis('equal')

plt.pie(exp_vals,labels=exp_labels, shadow=True)
plt.axis("equal")
plt.show()
```

```python
# Showing percentages for every pie. Specify radius to increase chart size

plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5)
plt.show()
```

```python
# Explode
# Which Pie to explode and by how much can be controlled by passing a list[] in explode arguement 

plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2])
plt.show()
```

```python
# Rotating Pie Chart by specifying the Start angle

plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2],counterclock=True, startangle=45)
plt.show()
```
